```python
#import json and store file in dictionary
import json
with open('exam2_tweets.json', 'r') as json_file:
    json_list = json.load(json_file)
json_list
```

```python
#each tweet is dictionary in list, heres tweet 1
print(json_list[0].keys())
print(json_list[0].values())
print(len(json_list[0].keys())) # of key value pairs
```

```python
#show one key value pair of a given key #id : 1698308935
tweet_1 = json_list[0]
print(tweet_1)
print(f"id : {tweet_1['id']}")
```

```python
#show one key value pair #id : 1698308935
tweet_1_keys = list(tweet_1.keys())
tweet_1_values = list(tweet_1.values())
print(tweet_1_keys[0], ":", tweet_1_values[0])
```

```python
#traverse the dictionary
for key in tweet_1.keys():
    print(key, tweet_1[key])
```
