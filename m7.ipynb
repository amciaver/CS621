```python
import pandas as pd
data = pd.read_csv("video_rental.csv")

#1) How many rows and column does this dataset have. 
# 40 rows 4 col
df = pd.DataFrame(data)
df.head(5)
```

```python
#2) How many male and female customers in this dataset? 
# 21,19
grouped = df['Income'].groupby(df['Gender'])
grouped.size()
```

```python
#3) What is the average income of the female customers.
# around 40k
grouped.mean()
```

```python
#4) For male customers who love Drama genre, what is the oldest age?
#38
df.groupby(['Genre','Gender']).max()
```

```python
#5) What is the youngest age of male customers?
#16
df.groupby(['Gender']).min()
```

```python
#6)What is the average age among female customers who love Drama movies? 
#about 43
df.groupby(['Gender','Genre']).mean()
```

```python
#7) Use the Genre column to draw a bar chart. Take a screenshot of the visualization you created and upload the image file
import matplotlib.pyplot as plt
t = df['Genre'].value_counts()
x = list(t.index)
y = list(t.values)
plt.bar(x,y)
```

```python
#8) Question 8
#Use the Genre column to draw a pie chart.
#Each slice needs to show its percentage value. 

plt.pie(y,labels = x) #doesnt show percent
```

```python
customers = df.groupby(['Gender','Genre'])['Age'].sum().unstack()
customers.plot(kind='bar',stacked=True)
plt.title('Age by Genre and Gender')
```
