```python
import pandas as pd
data = pd.read_csv("amazon_10_years.csv", parse_dates=True, index_col=0)
data
```

```python
df = pd.DataFrame(data)
df
```

```python
#which date hhas the largest volume number
df.groupby(['Volume','Date']).max()
```

```python
#which date has the lowest open value number 
df.groupby(['Open','Date']).max()
```

```python
#Which date has the largest positive difference between open and close prices? 
#(close price is higher than open price)
df.loc[(df['Close']-df['Open']).idxmax(),'Date']
```

```python
#Which date has the largest negative difference between open and close prices? 
#(close price is lower than open price) 
#or 
#df.loc[(df['Close']-df['Open']).idxmin(),'Date']
df.loc[(df['Open']-df['Close']).idxmax(),'Date']
```

```python
#5.Question 5 Which year has the largest total volume number?
max_volume = df.loc[df['Volume'].idxmax(),'Date']
max_volume
```

```python
type(max_volume)
```

```python
from datetime import datetime
splitted = max_volume.split("-")
splitted
```

```python
date = datetime(int(splitted[0]),int(splitted[1]),int(splitted[2]))
date
```

```python
date.year
```

```python
#6. Question 6 Which weekday has the highest average volume number?
#Monday is 0
weekday_averages = df.groupby(df.index.weekday)['Volume'].mean()
weekday_averages
```

```python
# 7. Question 7 Create a time series line chart using Date and Close columns.
df_close = df[['Close']]
df_close.head(5)
```

```python
df_close.plot.line(y='Close')
```

```python
#8.0 Create a time series line chart using Date and Close columns, 
# Also add another line using rolling mean with window size 250. 
df['Moving Average'] = df['Close'].rolling(window=250).mean()
df_close.plot.line(y=['Close','Moving Average'])
```

```python
#9.0 Create a time series line chart using Date and Close column.
#Also add another line using rolling std with window size 250. 
df['Rolling Std'] = df['Close'].rolling(window=250).std()
df['Rolling Std']
```

```python
df_close.plot.line(y=['Close','Rolling Std'])
```
